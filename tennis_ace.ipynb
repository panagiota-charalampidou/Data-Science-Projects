```python
#Import the necessary packeges
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#load and inspect the data
df=pd.read_csv('tennis_stats.csv')
print(df.head())
```

## Exploratory Analysis

```python
plt.scatter(x=df.DoubleFaults, y=df.Wins)
plt.title('Correlation between number of times player missed both first and second serve attempts and number of matches won in a year')
plt.xlabel('Double Faults')
plt.ylabel('Wins')
plt.show()
```

```python
plt.scatter(x=df.ServiceGamesPlayed, y=df.Wins)
plt.title('Correlation between total number of games where the player served and number of matches won in a year')
plt.xlabel('Service')
plt.ylabel('Wins')
plt.show()
```

```python
plt.scatter(x=df.BreakPointsConverted, y=df.Winnings)
plt.title('Correlation between % of the time the player was able to win their opponent’s service game when they had the chance and total winnings in USD($) in a year')
plt.xlabel('Break Points')
plt.ylabel('Winnings')
plt.show()
```

## Simple Linear Regression

```python
X=df[['ServiceGamesPlayed']]
y=df[['Wins']]

x_train, x_test, y_train, y_test=train_test_split(X,y, train_size=0.8)

slr=LinearRegression()
slr.fit(x_train, y_train)
slr.score(x_test,y_test)


y_predict= slr.predict(x_test)


plt.scatter(y_test, y_predict, alpha=0.4)
plt.title('Observed vs Predicted Number of Wins')
plt.xlabel('Observed')
plt.ylabel('Predicted')
plt.show()
```

```python
X=df[['Ranking']]
y=df[['Winnings']]

x_train, x_test, y_train, y_test=train_test_split(X,y, train_size=0.8)

slr=LinearRegression()
slr.fit(x_train, y_train)
slr.score(x_test,y_test)


y_predict= slr.predict(x_test)


plt.scatter(y_test, y_predict, alpha=0.4)
plt.title('Observed vs Predicted Winnings')
plt.xlabel('Observed')
plt.ylabel('Predicted')
plt.show()
```
